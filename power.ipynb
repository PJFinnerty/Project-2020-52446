```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

x = pd.read_csv("powerproduction.txt")

power = x["power"]

speed = x["speed"]

x['ind'] = x.index
```

```python
def power(t, c0, c1, c2, c3):
    return c0+c1*t-c2*np.exp(-c3*t)
```

```python
g = [20, 0.5, 20, 1]

n = len(x['speed'])
y = np.empty(n)
for i in range(n):
    y[i] = power(x['speed'][i], g[0], g[1], g[2], g[3])
    

plt.plot(x['speed'], x['power'])
plt.plot(x['speed'], y, 'r')
```

#### Fit Using Curve Fit

```python
from scipy.optimize import curve_fit

t = x['speed'].values
hr = x['power'].values

# BPM = Function Name, t = object to predict BPM, hr = measured values,
# g = guess values
c, cov = curve_fit(power, t, hr, g)
print(c)
```

```python

n = len(x['speed'])
y = np.empty(n)
for i in range(n):
    y[i] = power(x['speed'][i], c[0], c[1], c[2], c[3])
    

plt.plot(x['speed'], x['power'])
plt.plot(x['speed'], y, 'r.')
```

```python
from sklearn.metrics import r2_score
print('R^2: ', r2_score(y, hr))
```
